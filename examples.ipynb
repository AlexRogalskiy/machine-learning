# Installation and Examples

Steps for setting up jupyter with octave:
1. Install the lastest version of Anaconda (with python 3)
2. Install Homebrew
3. `brew update`
4. `brew install octave`
5. Add the following  to a file called ~/.octaverc
    ```
    setenv('GNUTERM','qt')
    graphics_toolkit("gnuplot")
    ```
6. `conda config --add channels conda-forge `
7. `conda install octave_kernel`
8. `conda install texinfo # For the inline documentation (shift-tab) to appear.`

If all goes well we should see gnuplot when we run the command `available_graphics_toolkits` in octave.

First lets show some basics in octave.

````octave
% Initialize matrix A and B 
A = [1, 2, 4; 5, 3, 2]
B = [1, 3, 4; 1, 1, 1]

% Initialize a vector 
v = [1;2;3] 

% Initialize constant s 
s = 2

% Initialize a 2 by 2 identity matrix
I = eye(2)
````

````octave
% Get the dimension of the matrix A where m = rows and n = columns
[m,n] = size(A)

% You could also store it this way
dim_A = size(A)

% Get the dimension of the vector v 
dim_v = size(v)

% Now let's index into the 2nd row 3rd column of matrix A
A_23 = A(2,3)
````

````octave
% See how element-wise addition works
add_AB = A + B 

% See how element-wise subtraction works
sub_AB = A - B

% See how scalar multiplication works
mult_As = A * s

% Divide A by s
div_As = A / s

% What happens if we have a Matrix + scalar?
add_As = A + s

% Multiply A * v
Av = A * v

% Multiply I * A
mult_IA = I*A
````

And now we'll show some plotting examples.

````octave
%plot --format svg
t = linspace(0,6*pi,100);
plot(sin(t))
grid on
hold on
plot(cos(t), 'r')
````

````octave
%plot --format png
tx = ty = linspace (-8, 8, 41)';
[xx, yy] = meshgrid (tx, ty);
r = sqrt (xx .^ 2 + yy .^ 2) + eps;
tz = sin (r) ./ r;
mesh (tx, ty, tz);
````

````octave
%plot -f png -w 600

% butterworth filter, order 2, cutoff pi/2 radians
b = [0.292893218813452  0.585786437626905  0.292893218813452];
a = [1  0  0.171572875253810];
freqz(b, a, 32);
````

````octave
% small image
imshow(randn(100,100))

% let's change the data range
figure
imshow(randn(100,100), [-10 10])

% let's try automatic range and different size
figure
imshow(randn(100,200), [])
colormap('autumn')
````

````octave
% Create an RGB image with no title or labels
close all hidden
image = zeros(300,400,3);
image(1:100,:,1) = 0.9;  % Red
image(101:200,:,2) = rand(100,400);  % Green
image(201:300,:,3) = 0.9;  % Blue
imshow(image)
title('Hello')
````

````octave
% let's add a title - this produces full figure
imshow(rand(100,100))

% let's create an image next to a plot
figure
subplot(121)
plot([1,2,3])
subplot(122)
imshow(randn(100,100))
````
